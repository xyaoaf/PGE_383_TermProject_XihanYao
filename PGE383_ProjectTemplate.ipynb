### Load Data

The following workflow applies the .csv file "LST_samples_with_bands_and_indices.csv", a synthetic dataset generated by Xihan Yao using the Landsat images near Austin, TX. Those images were acquired by the same satellite in Summer 2025. Landsat is a publicly available data set and can be acquired through multiple chanels. Here is a brief explanation of how the dataset is generated. 



This data was made up. I use the Land surface temperature layer from Landsat 8 as the source to generate my data. I took 1000 samples in total. First 500 samples are completely random. The other 500 samples were taken near the hotspots, where the temperature values of the first 500 samples are high. 

Due to urban heat island effects, the urbanized are is hotter then the surrounding rural area. This is why the temperature in Austin is higher than its periferals. I assume that there are more people living in the urbanized area, and thus more people are participating in citizen science and collecting data from their home, which results in denser samples in urbanized area. 

This, thus, lead to a sample that is not completely random. There are no samples on water.



Following that I took the values of Landsat b2 to b9 bands at those locations. And calculate the a few vegetation indices.


We will work with the following features:

* **Land Surface Temperature** - or LST, a measurment of temperature. This is not exactly the same as air temperature, but here we use it as a proxy to air temperature.
* **B2 Blue** - reflectence of the blue band light in electromeganatic specturem
* **B3 Gree** - product of sonic velocity and rock density in unitsof $kg/m^2s*10^3$
